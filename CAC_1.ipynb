```python
# Define a dictionary of contents with subcategories and recommendations
contents = {
    "Movies": {
        "Action Movies": ["Die Hard", "Fast and Furious", "John Wick", "The Dark Knight", "Speed"],
        "Comedy Movies": ["The Hangover", "Superbad", "Mean Girls", "Zoolander", "21 Jump Street"],
        "Romantic Movies": ["Me Before You", "La La Land", "Before Sunrise", "The Perks of being a Wallflower", "(500) Days of Summer" ],
        "Drama Movies": ["The Wolf of Wall Street", "Drunkrik", "The Shawshank Redemption" "Forrest Gump", "The Godfather"],
        "Science Fiction Movies": ["Tenet", "Blade Runner", "Dune", "The Matrix", "Interstellar"],
    },
    "Books": {
        "Fiction": ["The Hunger Games", "Ulysses" "To Kill a Mockingbird", "The Alchemist", "The Great Gatsby"],
        "Non-Fiction": ["In Cold Blood", "Sapiens", "Ariel", "Becoming", "Educated"],
        "Mystery": ["Gone Girl", "The Da Vinci Code" "The Girl on the Train", "Sherlock Holmes", "Runner"],
        "Science Fiction": ["Dune", "Ender's Game", "Station Eleven", "The Time Machine", "A Wrinkle in Time"],
        "Fantasy": ["Harry Potter", "Assassin's Apprentice", "The Lord of the Rings", "A Song of Ice and Fire", "Stardust"],
    },
    "Music": {
        "Rock": ["Jay-Z", "Carl Perkins", "Red Hot Chili Peppers", "The Beatles", "The Rolling Stones"],
        "Pop": ["Shawn Mendes", "V", "Taylor Swift", "Ariana Grande", "Harry Styles"],
        "Jazz": ["Norah Jones", "Ornette Coleman", "John Coltrane", "Charles Lloyd" "Ella Fitzgerald"],
        "Hip hop": ["Kanye West", "Cardi B", "Drake", "Imagine Dragons", "Maroon 5"],
        "Classical": ["Coldplay", "Mozart", "Bach", "The Rolling Stones", "The 1975"],
    },
    "TV Shows": {
        "Drama Series": ["The Walking Dead", "Game of Thrones", "The Witcher", "Stranger Things", "Breaking Bad"],
        "Comedy Series": ["How I Met Your Mother", "Friends", "The Office", "Brooklyn Nine-Nine", "Modern Family"],
        "Crime Series": ["The Boys", "Sherlock", "Daredevil", "Mindhunter", "Wednesday", "The Walking Dead"],
        "Fantasy Series": ["The Sandman", "The Witcher", "The Lord of the Rings", "The Mandalorian", "Alice in Borderland"],
        "Animation": ["The Simpsons", "Rick and Morty", "Attack on Titan", "SpongeBob SquarePants", "Avatar: The Last Airbender"]
    },
    "Video Games": {
        "Action Games": ["God of War", "Call of Duty", "GTA V", "Marvel's Spider-Man: Miles Morales", "Fortnite"],
        "Adventure Games": ["The Last of Us", "The Legend of Zelda", "Stray", "Uncharted", "Red Dead Redemption"],
        "RPG Games": ["Dragon Quest XI", "The Elder Scrolls V: Skyrim", "The Witcher 3", "Final Fantasy VII", "Darkest Dungeon"],
        "Strategy Games": ["Civilization VI", "StarCraft", "Stellaris", "Total War: Three Kingdoms", "Civilization VI"],
        "Mobile Games": ["Ludo King", "Clash of Clans", "FIFA Mobile", "Candy Crush Saga", "Subway Surfers"]
    },
    "Food": {
        "Italian Cuisine": ["Pizza Margherita", "itello Tonnato", "Pasta Carbonara", "Gnocchi di Patate", "Tiramisu"],
        "Mexican Cuisine": ["Tamales", "Tacos", "Guacamole", "Enchiladas", "Tacos al pastor"],
        "Chinese Cuisine": ["Kung Pao Chicken", "Spring Rolls", "Dim Sum", "Mapo Tofu", "Peking Duck"],
        "Indian Cuisine": ["Chicken Tikka Masala", "Masala Dosa", "Naan", "Biryani", "Fish curry"],
    },
    "Travel": {
        "Beach Destinations": ["Maldives", "Pondicherry", "Hawaii", "Goa", "Rio de Janeiro"],
        "Mountain Retreats": ["Swiss Alps", "Wildflower Hall", "Rocky Mountains", "Himalayas", "Bensley Collection"],
        "City Breaks": ["Paris", "New York City", "Tokyo", "Chennai", "Manchester"],
        "Cultural Experiences": ["Rome", "Kyoto", "Machu Picchu", "Coimbatore", "Mexico"],
    },
}

# Create or open a file for appending recommendations
def append_recommendations_to_file(category, subcategory, recommendations):
    filename = f"{category}_{subcategory}_recommendations.txt"
    with open(filename, "a") as file:
        for recommendation in recommendations:
            file.write(f"- {recommendation}\n")

# Display a menu for categories
def display_menu():
    print("Categories:")
    for index, category in enumerate(contents.keys(), 1):
        print(f"{index}. {category}")
    print(f"{len(contents) + 1}. Exit")

# Main program loop
while True:
    display_menu()
    choice = input("Select a category (or 'exit' to quit): ")

    if choice == 'exit':
        break

    try:
        choice = int(choice)
        if 1 <= choice <= len(contents):
            selected_category = list(contents.keys())[choice - 1]
            subcategories = contents[selected_category]
            print(f"Subcategories for {selected_category}:")
            for index, subcategory in enumerate(subcategories.keys(), 1):
                print(f"{index}. {subcategory}")
            print(f"{len(subcategories) + 1}. Back to Categories")
            sub_choice = input("Select a subcategory or enter to go back: ")

            if sub_choice == '':
                continue

            try:
                sub_choice = int(sub_choice)
                if 1 <= sub_choice <= len(subcategories):
                    selected_subcategory = list(subcategories.keys())[sub_choice - 1]
                    recommendations = subcategories[selected_subcategory]
                    if recommendations:
                        print("Recommendations:")
                        for recommendation in recommendations:
                            print(f" - {recommendation}")
                        append_recommendations_to_file(selected_category, selected_subcategory, recommendations)
                        print("Recommendations appended to file.")
                    else:
                        print("No recommendations available for this subcategory.")
                elif sub_choice == len(subcategories) + 1:
                    continue
                else:
                    print("Invalid subcategory choice.")
            except ValueError:
                print("Invalid input. Please enter a number for subcategory choice.")
        elif choice == len(contents) + 1:
            break
        else:
            print("Invalid category choice.")
    except ValueError:
        print("Invalid input. Please enter a number for category choice.")
```
