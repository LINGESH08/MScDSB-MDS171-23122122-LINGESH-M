```python
import random

items = ['Scale' , 'Laptop' , 'Chicken' , 'Paper' , 'Book' , 'Milk' , 'Cake' , 'Mobile' , 'Pen' , 'Coffee']
quant = ['1' , '4' , '2' , '9' , '6' , '11' , '16' , '6' , '8' , '2']
price = ['10' , '70' , '86' , '45' , '93' , '13' , '75' , '49' , '39' , '99']

i=1
while i <= 100:
    ran1 = random.randint(0,9)
    ran2 = random.randint(0,9)
    ran3 = random.randint(0,9)

    print("Random Index from Items List: ",ran1,'\n')
    print("Random Index from Qty List",ran2,'\n')
    print("Random Index from Price List",ran3,'\n')

    item = items[ran1]
    qr = quant[ran2]
    pr = price[ran3]

    qn = str(quant)
    cost = str(pr)
    com = item+','+qn+','+cost
    print(com,'\n')

    file = open('23122122.csv','a+')
    file.write(com)
    file.write('\n')
    file.close()

    i = i + 1
```
